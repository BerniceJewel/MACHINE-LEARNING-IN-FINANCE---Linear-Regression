```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
from ISLP.models import (ModelSpec as MS, summarize)
```

```python
from ISLP import load_data
Carseats = load_data('Carseats')
Carseats.columns
```

```python
print(Carseats.head())
```

```python
Carseats = Carseats.astype({'ShelveLoc': 'category', 'Urban': 'category', 'US': 'category'})
```

```python
Carseats.dtypes
```

 **a.   Fit a multiple regression model to predict **Sales** using ***Price***,
 ***Urban***, and ***US***.**


```python
#Fit the model

design = MS(['Price', 'Urban', 'US'])
X = design.fit_transform(Carseats)
y = Carseats['Sales']
```

```python
model = sm.OLS(y, X)
results = model.fit()
summarize(results)
```

**b.   Provide an interpretation of each coefficient in the model. Be careful—some of the variables in the model are qualitative!**





1. The intercept **13.0435** represents the number of car seats sold on average when all other predictors are disregarded.
   
2. The `Price` coefficient is negative and so sales will fall by roughly 545 seats(0.0545x1000)for every unit($1) increase in price.

3. The `Urban=Yes` coeff is not **statistically significant**.Stores in urban locations sell 0.0219 fewer units (in thousands) than average. However we note the p-value is very high **(0.936)**indicating there is no statistical evidence of a difference in average units sold between urban and rural stores.
   
4. The `US=Yes` coeff is 1.2, and this means an average increase in car seat sales of (1.2006x1000)=12006 units when `US=Yes`


 **c. Write out the model in equation form, being careful to handle
 the qualitative variables properly.**

**Linear regression equation**




$$y = 13.0435\ + -0.0545X_2 \ + -0.0219X_2 \ + 1.20006X_3$$


$$X_1\ =Price ,\\ X_2\ =Urban[YES] \ and \ 0\ if\ Rural, \ 1\  if\ Urban,  \\  X_3\ = US[YES]  \ 0\ if\  Store \ is\  Not\  in \ US,  \ 1\  if\ it\  is\  in\  US.$$






**d.   For which of the predictors can you reject the null hypothesis
 $$H\ _0 : β\ _ j =0$$**


We can reject the null hypothesis for the predictors Price and US[Yes].
This implies that their effect on sales is statistically significant.
